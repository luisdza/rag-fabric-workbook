# Environment Setup
To proceed with this workshop, you must create a Lakehouse within your Microsoft Fabric workspace and set up the necessary resources in your Azure account. Below, we'll outline the steps to accomplish this.

## Azure Setup
To complete this workshop, you'll need an Azure account. If you don't have one, you can easily create a free account before starting.

Make sure that the subscription you're using has the necessary permissions to create and manage resources.

Here's how to set up **Azure AI Search**:
1. Go to the [Azure Portal](https://portal.azure.com/) and click on `Create a resource`.
2. Search for `Azure AI Search` and select it.
3. Click on `Create.`
4. In the creation wizard, create a new resource group or select an existing one.
5. To minimise costs, change the pricing tier from "Standard" to "Free."
6. Click `Review + Create` and once the validation is successful, click `Create.`
![Azure AI Search](https://github.com/luisdza/rag-fabric-workbook/raw/main/images/azure-ai-search.png)

Similarly, follow these steps to set up **Azure AI services**:
1. Go to the [Azure Portal](https://portal.azure.com/) and click on `Create a resource.`
2. Search for `Azure AI services` and select it.
3. Click on `Create`.
4. In the creation wizard, choose the same resource group used for Azure AI Search.
5. Choose the "Standard S0" pricing tier.
6. Click `Review + Create` and once the validation is successful, click `Create.`
![Azure AI Document Intelligence](https://github.com/luisdza/rag-fabric-workbook/raw/main/images/azure-ai-services.png)

## Azure OpenAI Setup
You can skip this step if your Microsoft Fabric has an F64 or higher SKU.

However, if you're using the trial version of Microsoft Fabric or it doesn't have an F64+ SKU, you'll need to request access to the Azure OpenAI API using this form. Once you gain access to the Azure OpenAI API, follow these steps to create the Azure OpenAI resource in the Azure Portal:

1. Go to the [Azure Portal](https://portal.azure.com/) and click on `Create a resource`.
2. Search for `Azure OpenAI` and select the Azure OpenAI resource.
3. Click on `Create` to begin the creation process.
4. Fill out all the required fields in the creation wizard and click `Review + Create`.
5. Once the validation is successful, click `Create`. Select the same Resource Group as the Azure AI Search and Azure AI Services resources.
![Azure OpenAI](https://github.com/luisdza/rag-fabric-workbook/raw/main/images/azure-openai.png)

Next, you'll need to create new model deployments. Here's how:

1. Navigate to `Azure OpenAI Studio`.
2. Under management, click on `Deployments`.
3. Click on `Create Deployment`.
4. Create two deployments: one for the `text-embedding-ada-002` model and another for the `gpt-35-turbo-16k` model.

For each deployment, provide the keys and deployment names for the Azure OpenAI resource in the sections that are utilising the Azure OpenAI models.
![Azure OpenAI Deploy Model](https://github.com/luisdza/rag-fabric-workbook/raw/main/images/azure-openai-deploy-model.png)

## Create a Lakehouse
To set up a new Lakehouse in your Microsoft Fabric workspace, start by opening the Synapse Data Engineering interface. Then, click on the `Lakehouse` button. Enter `rag_workshop` as the name and click on `Create`.
![Fabric Lakehouse](https://github.com/luisdza/rag-fabric-workbook/raw/main/images/fabric-lakehouse.png)
