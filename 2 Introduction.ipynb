# Introduction
This workshop addresses the challenges and opportunities presented by the vast documents companies accumulate over time. Our primary focus will be developing a Retrieval-Augmented Generation (RAG) on Microsoft Fabric, leveraging Azure AI Document Intelligence, Azure AI Search, and Azure OpenAI. This system aims to produce responses based on a **company's internal documents**, providing a practical application for generative AI technologies like ChatGPT.

Generative AI models, while proficient in generating responses on a wide range of topics, are not infallible. They sometimes produce **"hallucinations"** - responses that seem accurate but are not. Therefore, our workshop's crucial aspect will be exploring grounding techniques. Grounding ensures that answers are based on verifiable sources, enabling users to validate the accuracy of AI-generated information. This approach mitigates the inherent limitations of generative AI models, which do not deliver perfectly accurate answers.

By guiding participants through building an RAG that utilises company-specific documents to generate responses, we aim to enhance the utility and reliability of generative AI within the business context. This practical application demonstrates the capability to customise AI responses and emphasises the importance of **transparency and user verification** in AI-generated content.
